# Data Visualization

```python
# Distribution of Job Roles
job_role_counts = df['JobRole'].value_counts()
print(job_role_counts)

# Bar chart for categorical variable
plt.bar(job_role_counts.index, job_role_counts, color='skyblue')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.title('Distribution of Job Roles')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Distribution of Monthly Income
plt.hist(df['MonthlyIncome'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.title('Distribution of Monthly Income')
plt.show()
```

```python
# Distribution of Attrition
attrition_counts = df['Attrition'].value_counts()
print(attrition_counts)
# Bar chart for categorical variable
plt.bar(attrition_counts.index, attrition_counts, color=['green', 'red'])
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.title('Distribution of Attrition')
plt.show()
```

```python
# Scatter plot for Age vs. Years at Company
plt.scatter(df['Age'], df['YearsAtCompany'], color='purple', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Years at Company')
plt.title('Relationship between Age and Years at Company')
plt.show()
```

```python
# Distribution of Marital Status
marital_counts = df['MaritalStatus'].value_counts()
print(marital_counts)
# Bar chart for categorical variable
plt.bar(marital_counts.index, marital_counts, color=['orange', 'blue', 'green'])
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Distribution of Marital Status')
plt.show()
```

```python
# Distribution of Job Levels
job_level_counts = df['JobLevel'].value_counts()
print(job_level_counts)
# Bar chart for categorical variable
plt.bar(job_level_counts.index, job_level_counts, color='lightgreen')
plt.xlabel('Job Level')
plt.ylabel('Count')
plt.title('Distribution of Job Levels')
plt.show()
```

```python
# Scatter plot for Years at Company vs. Monthly Income
plt.scatter(df['YearsAtCompany'], df['MonthlyIncome'], color='purple', alpha=0.5)
plt.xlabel('Years at Company')
plt.ylabel('Monthly Income')
plt.title('Relationship between Years at Company and Monthly Income')
plt.show()
```

```python
# Distribution of Percent Salary Hike
plt.hist(df['PercentSalaryHike'], bins=15, color='gold', edgecolor='black')
plt.xlabel('Percent Salary Hike')
plt.ylabel('Frequency')
plt.title('Distribution of Percent Salary Hike')
plt.show()
```

```python
# Box plot for Monthly Income by Job Role
plt.figure(figsize=(12, 8))
sns.boxplot(x='JobRole', y='MonthlyIncome', data=df, palette='Set3')
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.title('Monthly Income Distribution by Job Role')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```
